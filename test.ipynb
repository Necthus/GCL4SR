```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# 假设 edge_index 是一个形状为 (2, num_edges) 的张量
edge_index = np.array([[0,0,1,2,2,3,3,3,4,4,5,6],[2,3,2,4,5,2,4,5,5,6,1,2]])

# 创建一个无向图
G = nx.Graph()

# 添加节点
num_nodes = np.max(edge_index) + 1
G.add_nodes_from(range(num_nodes))

# 添加边
for i in range(edge_index.shape[1]):
    G.add_edge(edge_index[0, i], edge_index[1, i])

# 绘制图
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', font_size=10)
plt.show()
```

```python
import torch
from torch_geometric.loader import NeighborSampler
edge_index=torch.tensor([[0,0,1,2,2,3,3,3,4,4,5,6],[2,3,2,4,5,2,4,5,5,6,1,2]])
train_loader = NeighborSampler(edge_index, node_idx=torch.tensor([0,1,2,3],dtype=torch.long),sizes=[3,2],batch_size=2,shuffle=False)
for batch_size, n_id, adjs in train_loader:
    print("batch_size",batch_size)
    print("n_id",n_id)
    print("adjs",adjs)
```
